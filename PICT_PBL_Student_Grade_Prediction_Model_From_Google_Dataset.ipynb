```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_csv("Google_Student_Data.csv")
df
```

```python
df.isnull().sum()
```

```python
categorical_data = ['sex',
 'address',
 'famsize',
 'reason',
 'schoolsup',
 'paid',
 'activities',
 'higher',
 'internet']
```

```python
dummies=pd.get_dummies(df[categorical_data])
dummies
```

```python
droppers = ['sex_F',
 'address_R',
 'famsize_GT3',
 'reason_other',
 'schoolsup_no',
 'paid_no',
 'activities_no',
 'higher_no',
 'internet_no']
```

```python
good_dummies = dummies.drop(droppers,axis=1)
good_dummies
```

```python
merged=pd.concat([df,good_dummies],axis=1)
merged
```

```python
ONE=merged.drop(categorical_data,axis=1)
ONE
```

```python
sum(ONE.isnull().sum())
```

```python
y = ONE["G3"]
y.shape
```

```python
x = ONE.drop(["G3"],axis=1)
x.shape
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.linear_model import ElasticNet as EN
from sklearn.linear_model import Lasso as LA
from sklearn.linear_model import LogisticRegression as LGR
from sklearn.ensemble import RandomForestRegressor as RF
from sklearn.naive_bayes import GaussianNB as GNB
from sklearn.neighbors import KNeighborsClassifier as KNC
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor as DTR
models = [LR,EN,LA,LGR,RF,GNB,KNC,SVR,DTR]
for i in range(len(models)):
    models[i] = models[i]()
print(models)
```

```python
from sklearn.model_selection import train_test_split
print("Training started....")
result = {}
for i in range(10):
    print(i*10,"% done...")
    X_train, X_test, y_train, y_test = train_test_split(x,y,train_size=0.8,random_state=i)
    for model in models:
        print(str(i*10+models.index(model)+1),"% done training with",model,end=" ")
        model.fit(X_train,y_train)
        s = model.score(X_test,y_test)
        print("score =",s)
        try:
            result[model]+=s*10
            continue
        except:
            pass
        result[model]=s*10
print(result)
```

```python
sorted_result = sorted(result.items(), key = lambda kv:(kv[1], kv[0]),reverse = True)
sorted_result
```
